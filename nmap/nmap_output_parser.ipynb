```python
import xml.etree.ElementTree as ET

def read_file(fname: str)-> str:
    with open(f"{fname}") as f:
        return f.read()

def parse_nmaprun_xml(fname):
    xml_tree = ET.parse(fname)
    root = xml_tree.getroot()
    outports = []
    for child in root:
        nmaprun_tag = child.tag
        nmaprun_attrib = child.attrib
        for nmap_child in root.iter(child.tag):
            if nmap_child.tag == "host":
                for elem in nmap_child:
                    if elem.tag == "ports":
                        for ports in elem:
                            if ports.tag == "port":
                                for port in ports:
                                    port = {
                                        "port_id": ports.attrib["portid"],
                                        "protocol": ports.attrib["protocol"],
                                        "service_name": port.attrib["name"] if "name" in port.attrib else None}
                                outports.append(port)
    return outports
```

```python
import pandas as pd
fname = "portscan_out.xml"
output = parse_nmaprun_xml(fname)
df = pd.DataFrame(output)
df
```
